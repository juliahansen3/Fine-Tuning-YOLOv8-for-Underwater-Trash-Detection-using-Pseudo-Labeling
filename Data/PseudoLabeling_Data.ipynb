```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Result 1, Confidence Threshold:

```python
# Confidence interval test 1 data:
CI_data = {"Confidence_Threshold": [50, 60, 70, 80, 90], "Precision": [0.861, 0.874, 0.884, 0.853, 0.856],
           "Recall": [0.85, 0.821, 0.828, 0.837, 0.846],
           "mAP (50)":[0.885, 0.881, 0.884, 0.884, 0.885], "mAP (50-95)": [0.665, 0.665, 0.669, 0.667, 0.667]}
CI_df= pd.DataFrame(CI_data)
CI_df.describe()
```

```python
CI_df
```

```python
plt.plot(CI_df.iloc[:, 0], CI_df.iloc[:, 1:], label= ["Precision", "Recall", "mAP (50)", "mAP (50-95)"]);
plt.legend()
plt.title("Change in Performance Metrics Over a Range of Confidence Thresholds")
plt.xlabel("Confidence Threshold (%)")
```
